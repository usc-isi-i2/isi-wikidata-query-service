```python
from faiss_similarity import FAISSIndex
import pandas as pd
from IPython.display import display, HTML
```

```python
large_all = '/Users/amandeep/Github/isi-wikidata-query-service/iwfs/test_data/text-embeddings'
wiki_labels = '/Users/amandeep/Github/isi-wikidata-query-service/iwfs/test_data/labels/labels.en.tsv.gz'
```

```python
import time
s = time.time()
fi = FAISSIndex(large_all, wiki_labels)
fi.build_index()
print(time.time() - s)
```

```python
queries = ['Q2452129', 'Q10422405']
df_list = []
for q in queries:
    df_list.append(pd.DataFrame(fi.nearest_neighbors(q, k=10)))
df = pd.concat(df_list)
display(HTML(df.to_html()))
```
